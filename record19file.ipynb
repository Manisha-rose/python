```python
#Python program to copy odd lines of one file to other 
with open("data.txt", "r") as src, open("otherdata.txt", "w") as dest:
 lines = src.readlines()
 for i in range(len(lines)):
  if i % 2 == 0: # odd line numbers (1,3,5,...)
   dest.write(lines[i])

print("Odd lines copied successfully!")
```

```python
#Write a Python program to read each row from a given csv file and print a list of strings. 
import csv

with open("data.csv","r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
#Write a Python program to read specific columns of a given CSV file and print the content of the columns. 
import csv 

with open("data.csv","r") as file:
    reader=csv.reader(file)
    for row in reader:
        name = row[0]   # Column 1 (index 0)
        rollno = row[1]   # Column 3 (index 2)
        print(name, rollno)
```

```python
#Write a Python program to write a Python dictionary to a csv file. After writing the CSV file read the CSV file and display the content.
import csv

data=[
    {"Name":"Anu","Age":22,"Place":"Calicut"},
    {"Name":"miya","Age":23,"Place":"Delhi"}
    
]
# Write dictionary to CSV
with open("students.csv", "w", newline='') as file:
 writer = csv.DictWriter(file, fieldnames=["Name", "Age", "Place"])
 writer.writeheader()
 writer.writerows(data)

# Read CSV back
with open("students.csv", "r") as file:
 reader = csv.reader(file)
 for row in reader:
   print(row) 
```
