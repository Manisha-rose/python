```python
class time:
    def __init__(self, hr, min, sec):
        self.h = hr
        self.m = min
        self.s = sec

    def __add__(self, other):
        hh = self.h + other.h
        mm = self.m + other.m
        ss = self.s + other.s
        
        
# Adjust seconds to minutes
        if ss >= 60:
          mm += ss // 60
          ss = ss % 60

# Adjust minutes to hours
        if mm >= 60:
           hh += mm // 60
           mm = mm % 60       
        return time(hh, mm, ss)

    def display(self):
        print("Hour:", self.h, "Minute:", self.m, "Second:", self.s)


h1 = int(input("Enter hour1: "))
m1 = int(input("Enter minute1: "))
s1 = int(input("Enter second1: "))
h2 = int(input("Enter hour2: "))
m2 = int(input("Enter minute2: "))
s2 = int(input("Enter second2: "))

# Create Time objects
t1 = time(h1, m1, s1)
t2 = time(h2, m2, s2)

# Add and display
result = t1 + t2
print("Added Time:")
result.display()
```
