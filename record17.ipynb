```python
#base class
class publisher:
    def __init__(self,name):
        self.name=name
    def display(self):
        print("The publisher name is:",self.name)

#derived class from publisher
class book(publisher):
    def __init__(self,name,title,author):
        super().__init__(name)
        self.title=title
        self.author=author
   #method overriding
    def display(self):
        super().display()
        print("The title of the book is:",self.title)
        print("Name of the author is:",self.author)

#derived class from book
class python(book):
    def __init__(self,name,title,author,price,no_of_pages):
        super().__init__(name,title,author)
        self.price=price
        self.no_of_pages=no_of_pages
   #method overriding 
    def display(self):
        super().display()
        print("The price of the book is:Rs.",self.price)
        print("No of pages in the book is:",self.no_of_pages)

print("Python Book Details:\n")

name=input("Enter publishers name:")
title=input("Enter title of the book:")
author=input("Enter author name:")
price=float(input("Enter price of the book:"))
no_of_pages=int(input("Enter no.of pages:"))
p1=python(name,title,author,price,no_of_pages)
print("\n__Python Book Information__")
p1.display()     
    
```
