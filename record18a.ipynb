```python
#parent class
class rectangle:
    def __init__(self,l,b):
      self.l=l
      self.b=b
      self.area=l*b

    def __lt__(self,other): # Overload the '<' operator
        if self.area<other.area:
            print("Rectagle 2 is greater")
        else:
            print("Rectangle 1 is greater")

l1=int(input("Enter length of rectangle1:"))
b1=int(input("Enter breadth of rectangle1:"))
l2=int(input("Enter length of rectangle2:"))
b2=int(input("Enter breath of rectangle2:"))

r1=rectangle(l1,b1)
r2=rectangle(l2,b2)
r1<r2
```
